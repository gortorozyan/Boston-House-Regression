```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import GradientBoostingRegressor
```

```python
house_data = pd.read_csv('house_data.csv')
house_data
```

```python
for i in house_data.mainroad.unique():
    col = f'mainroad_{i}'  
    house_data[col] = (house_data['mainroad'] == i).astype('int8')
    
for i in house_data.guestroom.unique():
    col = f'guestroom_{i}'  
    house_data[col] = (house_data['guestroom'] == i).astype('int8')
    
for i in house_data.basement.unique():
    col = f'basement_{i}'  
    house_data[col] = (house_data['basement'] == i).astype('int8')
    
for i in house_data.hotwaterheating.unique():
    col = f'hotwaterheating_{i}'  
    house_data[col] = (house_data['hotwaterheating'] == i).astype('int8')

for i in house_data.airconditioning.unique():
    col = f'airconditioning_{i}'  
    house_data[col] = (house_data['airconditioning'] == i).astype('int8')
    
for i in house_data.furnishingstatus.unique():
    col = f'furnishingstatus_{i}'  
    house_data[col] = (house_data['furnishingstatus'] == i).astype('int8')
```

```python
house_data.drop(['mainroad', 'guestroom', 'basement', 'hotwaterheating', 'airconditioning', 'furnishingstatus', 'prefarea'], axis=1, inplace=True)
```

```python
house_data
```

```python
x = house_data.drop('price', axis=1)
```

```python
y = house_data.price
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.6)
```

```python
model = GradientBoostingRegressor()
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```
